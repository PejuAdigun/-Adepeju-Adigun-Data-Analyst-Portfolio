Changing the release_date column to a datetime type while handling any errors that could occur. We need this for further analysis of how many releases per year

```python
netflix_dt['release_date'] = pd.to_datetime(netflix_dt['release_year'], format='%Y', errors='coerce')
```

```python
# Group by year and count
releases_per_year = netflix_dt['release_year'].value_counts().sort_index()
releases_py = releases_per_year.reset_index()
releases_py.columns = ["Year", "Number of Releases"]
release_py_df = releases_py.columns
```

```python
releases_py
```

```python
# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=releases_py,x="Year", y='Number of Releases', color='skyblue')
plt.title('Number of Netflix Releases per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

This is handling the genre column, to find out the distribution between each genre. By splitting it at each comma before exploding it into different rows, then counting each occurence of the genre.

```python
genre = netflix_dt["listed_in"].str.split(',')
genres_exploded = genre.explode()
genre_counts = genres_exploded.str.strip().value_counts()
genre_df = genre_counts.reset_index()
genre_df.columns = ['Genre', 'Count']
```

```python
# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=genre_df.head(20), x='Genre', y='Count', palette='viridis')
plt.xticks(rotation=75)
plt.title('Top 20 Netflix Genres')
plt.show()
```

This is where we join the strings in description into a single long string and changing the type to string for word cloud. We join it because we have multiple text entries (one per row). But the WordCloud algorithm works best when you give it one continuous block of text

```python
text = ' '.join(netflix_dt["description"].dropna().astype(str))
```

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, max_words=100).generate(text)

# Display the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud of Movie Descriptions')
plt.show()
```
